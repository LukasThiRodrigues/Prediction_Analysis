### Passo 1 - Importação

Essa parte é apenas a conexão com o banco de dados para enviar o dataset inicial.

```python
# Instalação de Dependências
%pip install azure-storage-blob
%pip install kagglehub

# Importação das Bibliotecas
import kagglehub
from azure.storage.blob import BlobServiceClient
from dotenv import load_dotenv
import shutil
import os

# Definição das Variáveis
load_dotenv("../.env")
connection_string = os.getenv("CONNECTION_STRING")
file_path = "../data/pokemon.csv"
dest_dir = "../data"
container_name = "raw"
blob_name = "pokemon.csv"

# Download do Dataset do Kaggle
src_dir = kagglehub.dataset_download("rounakbanik/pokemon")

os.makedirs(dest_dir, exist_ok=True)
for fname in os.listdir(src_dir):
    shutil.copy(os.path.join(src_dir, fname), os.path.join(dest_dir, fname))

# Upload do Dataset para o Azure
blob_service_client = BlobServiceClient.from_connection_string(connection_string)
blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)

with open(file_path, "rb") as data:
    blob_client.upload_blob(data, overwrite=True)
```
