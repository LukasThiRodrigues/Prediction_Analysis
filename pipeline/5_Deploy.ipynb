```python
import pandas as pd
import requests

# CSV com novos Pokémon
df_new = pd.read_csv("../data/novo_pokemon.csv")
url = "http://localhost:7071/api/predict_pokemon"

# Transformar DataFrame em lista de dicionários
data_payload = df_new.to_dict(orient="records")

# Chamada POST
response = requests.post(url, json=data_payload)

# Resultado
predictions = pd.DataFrame(response.json())
predictions.head()
```
